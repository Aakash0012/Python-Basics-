```python
class celsi:
    def __init__(self,a,b):
        self.a = a
        self.b = b
        
    def add(self):
        return self.a+self.b
    def sub(self):
        return self.a-self.b
    def multi(self):
        return self.a*self.b
    def div(self):
        return self.a/self.b
    def fact(self, n):
        if n==1:
            return 1
        else:
            return n*fact(n-1)
        
        
```

```python
obj = celsi(3,7)
obj
```

```python
a = obj.multi()
a
```

```python
b = obj.fact(n=3)
b
```

```python
from bs4 import BeautifulSoup
import requests

url = 'https://www.w3schools.com/html/html_tables.asp'
r = requests.get(url)
soup = BeautifulSoup(r.content, 'html.parser')

a = soup.findAll('table', attrs = {'id':'customers'})
ro = a[0].findAll('tr')

for i in ro[0].findAll('th'):
    print(i.text, end='\t')
        
print()
for j in range(1, len(ro)):
    for i in ro[j].findAll('td'):
        print(i.text, end='\t')
    print()
```
