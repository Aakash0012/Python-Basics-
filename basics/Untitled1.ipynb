```python
from bs4 import BeautifulSoup
import requests

url = 'https://www.w3schools.com/html/html_tables.asp'
r = requests.get(url)
soup = BeautifulSoup(r.content, 'html.parser')
soup
```

```python
a= soup.findAll('table', attrs = {'id':'customers'})
a
```

```python
ro = a[0].findAll('tr')
ro
```

```python
for i in ro[0].findAll('th'):
    print(i.text, end='\t')
    print()
```

```python
for j in range(1, len(ro)):
    for i in ro[j].findAll('td'):
        print(i.text, end='\t')
        print()+
```
